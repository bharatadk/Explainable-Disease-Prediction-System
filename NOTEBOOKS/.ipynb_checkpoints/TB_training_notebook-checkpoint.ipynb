```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
from tensorflow.keras import models,layers
from tensorflow.keras.preprocessing import image
from tensorflow.keras.optimizers import Adam
```

```python
IMAGE_SIZE = 224
BATCH_SIZE = 32
CHANNELS = 3
EPOCHS = 50
```

```python
def to_grayscale_then_rgb(image):
    image = tensorflow.image.rgb_to_grayscale(image)
    image = tensorflow.image.grayscale_to_rgb(image)
    return image
```

```python
train_datagenerator = image.ImageDataGenerator(
    preprocessing_function=to_grayscale_then_rgb
)
val_datagenerator = image.ImageDataGenerator(
    preprocessing_function=to_grayscale_then_rgb
)
```

```python
train_dataset =train_datagenerator.flow_from_directory(
    "/content/drive/MyDrive/TB_Datasets_new/train",
    shuffle = True,
    target_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = BATCH_SIZE
)

val_dataset =val_datagenerator.flow_from_directory(
    "/content/drive/MyDrive/TB_Datasets_new/val",
    shuffle = True,
    target_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = BATCH_SIZE
)
```

```python
path_normal = "/content/drive/MyDrive/TB_Datasets_final/train/Normal"
path_tb = "/content/drive/MyDrive/TB_Datasets_final/train/Tuberculosis"
```

```python
import os
import cv2
import tensorflow
normal_image_names=[]
count = 0
for i in os.listdir(path_normal):
  normal_image_names.append(path_normal+'/'+i)
  count += 1
  if count == 6:
    break

print(normal_image_names)
all_images = [cv2.imread(image) for image in normal_image_names]
rows= 2
columns = 3
fig =plt.figure(figsize = (10,10))
for j in range(rows*columns):
  fig.add_subplot(rows,columns,j+1)
  plt.imshow(all_images[j])
  plt.title('Normal')
```

```python
tb_image_names=[]
count = 0
for i in os.listdir(path_tb):
  tb_image_names.append(path_tb+'/'+i)
  count += 1
  if count == 6:
    break

print(tb_image_names)
all_images = [cv2.imread(image) for image in tb_image_names]
rows= 2
columns = 3
fig =plt.figure(figsize = (10,10))
for j in range(rows*columns):
  fig.add_subplot(rows,columns,j+1)
  plt.imshow(all_images[j],cmap='gray')
  plt.title('Tuberculosis')
```

```python
import os
def count_images(path):
  count = 0
  for image_name  in os.listdir(path):
    count+=1
  return count
```

```python
cn = count_images(  path_normal)
print('Normal images count : ',cn)

ct=count_images(  path_tb)
print('Tuberculosis images count : ',ct)
```

```python
plt.bar(['Normal','TB'],[cn,ct])
plt.plot()
```


## **Since the data is unbalanced , so we try to balance the weightage of data**

```python
total = cn+ct
weight_for_0 = (1 / cn ) * (total/2.0)
weight_for_1 = (1/ct) * (total /2.0)
class_weight = { 0: weight_for_0, 1:weight_for_1}
print(class_weight)
```

```python
resize_and_rescale= tf.keras.Sequential([
                     layers.experimental.preprocessing.Resizing(IMAGE_SIZE,IMAGE_SIZE),
                     layers.experimental.preprocessing.Rescaling(1.0/255)
                    
])
```

```python
data_augmentation =tf.keras.Sequential([
                                        layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
                                        layers.experimental.preprocessing.RandomRotation(0.2),
])
```

# Making a CNN model

```python
input_shape= (BATCH_SIZE, IMAGE_SIZE,IMAGE_SIZE,CHANNELS)
model = models.Sequential([
                           resize_and_rescale,
                           data_augmentation,
                           layers.Conv2D(32,kernel_size = (3,3),activation = 'relu',input_shape = (IMAGE_SIZE,IMAGE_SIZE)),
                           layers.MaxPooling2D((2,2)),
                           layers.Conv2D(64,kernel_size =(3,3),activation = 'relu',padding = 'same'),
                           layers.MaxPooling2D((2,2)),
                           layers.Conv2D(64,kernel_size =(3,3),activation = 'relu',padding = 'same'),
                           layers.MaxPooling2D((2,2)),
                           layers.Conv2D(64,kernel_size =(3,3),activation = 'relu',padding = 'same'),
                           layers.MaxPooling2D((2,2)),
                           layers.Conv2D(64,kernel_size =(3,3),activation = 'relu',padding = 'same'),
                           layers.MaxPooling2D((2,2)),   
                           layers.Conv2D(64,kernel_size =(3,3),activation = 'relu',padding = 'same'),
                           layers.MaxPooling2D((2,2)),
                           layers.Flatten(),
                           layers.Dense(64,activation = 'relu'),
                           layers.Dense(2, activation = 'softmax'),                                                                
])

```

```python
model.build(input_shape= input_shape)
```

```python
model.summary()
```

```python
opt = Adam(lr=0.0001)

model.compile(
    optimizer = opt,
    loss= tf.keras.losses.CategoricalCrossentropy(),
    metrics = ['accuracy']
)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

model_save_path = '/content/drive/MyDrive/cnn_model_may.h5'
checkpoint = ModelCheckpoint(filepath=model_save_path,monitor= 'val_accuracy',verbose = 1,save_best_only = True,mode = 'max')
callbacks_list = [checkpoint]
```

```python
history = model.fit(
    train_dataset,
    epochs = 50,
    batch_size = BATCH_SIZE,
    verbose= 1,
    validation_data = val_dataset,
    callbacks = callbacks_list,
    class_weight=class_weight

)
```

```python
model.save('modelname.h5')
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
plt.figure(figsize = (10,10))
plt.subplot(1,2,1)
plt.plot(range(EPOCHS),acc,label = 'Training accuracy')
plt.plot(range(EPOCHS),val_acc,label = 'Validation Accuracy')
plt.legend(loc= 'lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1,2,2)
plt.plot(range(EPOCHS),loss, label = 'Training Loss')
plt.plot(range(EPOCHS),val_loss,label = 'Validation Loss')
plt.legend(loc = 'upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
def check_threshold_value(threshold):
  path ="/content/drive/MyDrive/TB_Datasets_final/val/Normal"
  import os
  tn = 0
  fp = 0 
  for image_name  in os.listdir(path):
      img_path = "/content/drive/MyDrive/TB_Datasets_final/val/Normal/"+image_name
      img = image.load_img(img_path,target_size = (224,224,3))
      img_arr = image.img_to_array(img)
      img_expand_dims = np.expand_dims(img_arr,axis = 0)
      p = model.predict(img_expand_dims)
      if p[0][1] <= threshold :
          tn+=1
      else:
          fp+=1


  tp = 0
  fn = 0 
  path = "/content/drive/MyDrive/TB_Datasets_final/val/Tuberculosis"
  for image_name  in os.listdir(path):
      img_path =  "/content/drive/MyDrive/TB_Datasets_final/val/Tuberculosis/"+image_name
      img = image.load_img(img_path,target_size = (224,224,3))
      img_arr = image.img_to_array(img)
      img_expand_dims = np.expand_dims(img_arr,axis = 0)
      p = model.predict(img_expand_dims)
      if p[0][1] >threshold :
          tp+=1
      else:
          fn+=1

  print('threshold : ',threshold, 'accuracy : ',(tn+tp)/(tn+tp+fp+fn))
```

```python
check_threshold_value(0.3)
```

```python
path ="/content/drive/MyDrive/TB_Datasets_final/test/Normal"
import os
tn = 0
fp = 0 
for image_name  in os.listdir(path):
    img_path = "/content/drive/MyDrive/TB_Datasets_final/test/Normal/"+image_name
    img = image.load_img(img_path,target_size = (224,224,3))
    img_arr = image.img_to_array(img)
    img_expand_dims = np.expand_dims(img_arr,axis = 0)
    p = model.predict(img_expand_dims)
    if p[0][1] <= 0.3 :
        tn+=1
    else:
        fp+=1
print('tn',tn)
print('fp',fp)
```

```python
import os
tp = 0
fn = 0 
path = "/content/drive/MyDrive/TB_Datasets_final/test/Tuberculosis"
for image_name  in os.listdir(path):
    img_path =  "/content/drive/MyDrive/TB_Datasets_final/test/Tuberculosis/"+image_name
    img = image.load_img(img_path,target_size = (224,224,3))
    img_arr = image.img_to_array(img)
    img_expand_dims = np.expand_dims(img_arr,axis = 0)
    p = model.predict(img_expand_dims)
    if p[0][1] >0.3 :
        tp+=1
    else:
        fn+=1

print('tp',tp)
print('fn',fn)
```
